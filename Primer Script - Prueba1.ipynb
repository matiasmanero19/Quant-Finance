```python
# pip install PyPortfolioOpt==1.2.1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
from pandas_datareader import data as pdr
import PyPortfolioOPt as pypfopt
plt.style.use('seaborn-colorblind')
```

```python
# activos
tickers = ['INTC', 'AMD', 'QCOM']
prices = pdr.get_data_yahoo(tickers, start = '2015-01-01', end = dt.date.today())['Adj Close']
returns = prices.pct_change()
```

```python
plt.figure(figsize=(14,7))
for i in prices.columns.values:
    plt.plot(prices.index, np.log(prices[i]), lw = 2, alpha = 2, label = i)
plt.legend(loc='upper left',fontsize=12)
plt.ylabel('log price')
    
```
